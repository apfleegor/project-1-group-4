```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load dataset
happiness_df = pd.read_csv('Gallup2018.csv')
```

### basic data analysis

```python
happiness_df.head()
```

```python
# describe the dataset
happiness_df.describe()
```

```python
# look for the state with the lowest well-being index

min_state = happiness_df.sort_values('Well-Being Index', ascending=True).iloc[0, 0]
min_score = happiness_df.sort_values('Well-Being Index', ascending=True).iloc[0, 2]
print(f'The state with the lowest well-being index score is {min_state} with a score of {min_score}.')
```

```python
# look for the state with the highest well-being index

max_state = happiness_df.sort_values('Well-Being Index', ascending=False).iloc[0, 0]
max_score = happiness_df.sort_values('Well-Being Index', ascending=False).iloc[0, 2]
print(f'The state with the lowest well-being index score is {max_state} with a score of {max_score}.')
```
