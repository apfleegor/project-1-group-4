```python
print(f"The mean number of publicly available tennis courts for US is: {round(consolidated_df['count_public_court_locations'].mean(),2)}")
print(f"The standard deviation of the number of all publicly available tennis courts is: {round(consolidated_df['count_public_court_locations'].std(),2)}")

well_being = consolidated_df.iloc[:,2]
public_courts_availability =  consolidated_df.iloc[:,12]

correlation = st.pearsonr(well_being,public_courts_availability)

print(f"The correlation coeff is: {round(correlation[0],2)}")
```
